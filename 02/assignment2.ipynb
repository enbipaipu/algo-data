# 無向グラフの最短経路を求める関数

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
Node_List = [x for x in range(1, 10)]
EDGE_List = [
    (1, 2, 4),
    (1, 8, 9),
    (8, 2, 11),
    (2, 3, 8),
    (8, 9, 7),
    (8, 7, 1),
    (9, 7, 6),
    (9, 3, 2),
    (7, 6, 2),
    (3, 4, 7),
    (3, 6, 4),
    (6, 4, 14),
    (6, 5, 10),
    (4, 5, 9),
]

G = nx.Graph()
G.add_nodes_from(Node_List)
G.add_weighted_edges_from(EDGE_List)

# 最短経路を計算
start = 5
end = 1
# ダイクストラ法で最短経路とその重みを求める
shortest_path = nx.dijkstra_path(G, start, end)
shortest_path_weight = nx.dijkstra_path_length(G, start, end)

# 出力
print("Shortest Path:", shortest_path)
print("Weight:", shortest_path_weight)

# グラフ描画のための設定
pos = nx.spring_layout(G)  # ノードの位置を計算
nx.draw_networkx_nodes(G, pos, node_color="skyblue")  # ノードを描画
nx.draw_networkx_edges(G, pos, edge_color="gray")  # エッジを描画
nx.draw_networkx_labels(G, pos)  # ノードのラベルを描画
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))  # エッジの重みを表示

# 最短経路上のノードとエッジを強調表示
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color="lightgreen")
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="green", width=2)

plt.axis("off")  # 軸の非表示
plt.show()  # グラフの表示
```
